```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/processed/events_processed.csv")

df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
attendance_by_category = (
    df.groupby("category_name")["attendance_count"]
      .mean()
      .sort_values(ascending=False)
)

attendance_by_category.plot(
    kind="bar",
    figsize=(8,5),
    title="Average Attendance by Category"
)

plt.ylabel("Average Attendance")
plt.tight_layout()
plt.show()
```

```python
attendance_by_day = (
    df.groupby("day_of_week")["attendance_count"]
      .mean()
)

attendance_by_day.plot(
    kind="bar",
    figsize=(8,5),
    title="Average Attendance by Day"
)

plt.ylabel("Average Attendance")
plt.tight_layout()
plt.show()
```

```python
attendance_by_weather = (
    df.groupby("weather_condition")["attendance_count"]
      .mean()
)

attendance_by_weather.plot(
    kind="bar",
    figsize=(8,5),
    title="Average Attendance by Weather"
)

plt.ylabel("Average Attendance")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["venue_capacity"],
    df["attendance_count"]
)

plt.xlabel("Venue Capacity")
plt.ylabel("Attendance")
plt.title("Venue Capacity vs Attendance")

plt.tight_layout()
plt.show()
```
